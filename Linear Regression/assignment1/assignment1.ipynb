```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

```

```python
#Converting dataframes to numpy arrays
train_data = train_df.values
test_data = test_df.values

print(train_data.shape)
```

```python
#Use normal equation method to compute the optimal parameters for linear regression. Print the values of optimal parameters.
m = train_data.shape[0]
#column x of train data will be in float format
X = train_data[:,0].astype(int)
Y = train_data[:,1].astype(float)

A11 = m
A12 = np.sum(X)
A21 = A12
A22 = np.sum(X**2)

B11 = np.sum(Y)
B21 = np.sum(X*Y)

A = np.array([[A11,A12],[A21,A22]])
B = np.array([[B11],[B21]])

#Optimal parameters

theta = np.linalg.inv(A).dot(B)
print(theta)

#Plot the data points and the regression line
plt.scatter(X,Y)
plt.plot(X,theta[0]+theta[1]*X,color='red')
plt.show()
```

```python
X_test = test_data[:,0].astype(int)
Y_test = test_data[:,1].astype(float)

Y_prime = theta[0]+theta[1]*X_test

MSE = np.sum((Y_test-Y_prime)**2)/m

print(MSE)
```

```python
plt.scatter(X_test,Y_test,color='green',label='Test Data',marker='.',s=5)
plt.plot(X_test,Y_prime,color='red',marker='*',label='Regression Line')
plt.show()
```
